```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing import image
```

```python
# --- SETTING ---
IMG_SIZE = 224
MODEL_PATH = "trash_classifier.keras"   # ganti kalau beda lokasi

# URUTANNYA HARUS SAMA seperti waktu training
label_names = [
    "Anorganik",
    "B3",
    "Organik"
]
```

```python
model = tf.keras.models.load_model(MODEL_PATH)
```

```python
# --- PREDICT FUNCTION ---
def predict_image(img_path):
    # load + resize
    img = image.load_img(img_path, target_size=(IMG_SIZE, IMG_SIZE))

    # convert to array
    img_array = image.img_to_array(img)


    # tambah batch dimension
    img_array = np.expand_dims(img_array, axis=0)

    # predict
    preds = model.predict(img_array)
    
    # hasil prediksi
    class_idx = np.argmax(preds[0])
    class_label = label_names[class_idx]
    confidence = preds[0][class_idx]

    return class_label, confidence
```

```python
# --- TEST ---
img_path = "ano.jpg"  # ganti gambar kamu
label, conf = predict_image(img_path)

print("Predicted Class :", label)
print("Confidence      :", round(float(conf)*100, 2), "%")
```
